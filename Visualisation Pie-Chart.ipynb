```python
import matplotlib.pyplot as plt
```

### Level -1

```python
marks = [45,50,55,60]
plt.pie(marks)
plt.show()
```

### Level -2 | Add labels

```python
marks = [45,50,55,60]
labels = ['Hindi','English','Math','Science']
plt.pie(marks, labels = labels)
plt.show()
```

### Level-3 | Adding angles

```python
marks = [45,50,55,60]
labels = ['Hindi','English','Math','Science']
plt.pie(marks, labels = labels,startangle = 90)
plt.show()
```

### Level -4 | adding explode

```python
marks = [45,50,55,60]
labels = ['Hindi','English','Math','Science']

explodes = [.01,.01,.01,.1]
plt.pie(marks,labels = labels, explode = explodes)
plt.show()
```

### Level -5 | Adding shadows

```python
marks = [45,50,55,60]
labels = ['Hindi','English','Math','Science']

explodes = [.01,.01,.01,.1]
plt.pie(marks,labels = labels, explode = explodes, shadow = True)
plt.show()
```

### Level -6 | Adding specific colours

```python
marks = [45,50,55,60]
labels = ['Hindi','English','Math','Science']
colours = ['b','r','g','orange']

explodes = [.01,.01,.01,.1]
plt.pie(marks,labels = labels, explode = explodes, shadow = True,colors = colours)
plt.show()
```

### Level - 7 | add Legend

```python
marks = [45,50,55,60]
 

plt.pie(marks, explode = explodes, shadow = True,colors = colours)
plt.legend(['Hindi','English','Math','Science'])
plt.show()
```

### Level - 8 | title

```python
marks = [45,50,55,60]
 

plt.pie(marks, explode = explodes, shadow = True,colors = colours)
plt.legend(['Hindi','English','Math','Science'], title = 'Subjects')
plt.show()
```

### Level - 9 | Adding numbers

```python
marks = [45,50,55,60]
labels = ['Hindi','English','Math','Science']

plt.pie(marks, explode = explodes, shadow = True,labels = labels,autopct = '%1.2f%%' )

plt.show()
```

# 2. Scatter Plot

```python
girls = [88,90,91,89,95]
boys  = [78,87,88,83,90]
minimum = [0,0,0,0,0]
maximum = [100,100,100,100,100]

standard = [6,7,8,9,10]
```

```python
plt.scatter(standard,girls, edgecolors = 'orange', marker = 's', c = 'blue',linewidth = 2)
plt.scatter(standard,boys, edgecolors = 'orange', marker = '^', c = 'green',linewidth = 2)
plt.scatter(standard,minimum, c = 'white')
plt.scatter(standard,maximum, c = 'white')
plt.legend(['Girls','Boys'])
plt.xlabel('Class')
plt.ylabel('Marks')
plt.title('Marks distribution of boys and girls in diffrent classes')
plt.show()
```
