### Loading the JSON file

```python
fd = open('js.json','r')
txt = fd.read()
fd.close()
```

### 2. Converting String to Dictionary

```python
import json
dct = json.loads(txt)
```

### 4. Convert Dictionary to List (NoSQL to SQL)

```python
data = []
for i in dct['data']:
#     print(i)
    data.append([i['id'],i['infoUrl'],i['name'],i['type']['id'],i['type']['name']])
```

### 5. Converting Multi- Dimensional List to csv

```python
import pandas as pd

df = pd.DataFrame(data, columns = ['id','url','name','type_id','type_name'])
```

```python
df.head()
```

### 6. Checking Null values in CSV file

```python
df.isnull().sum()
```

### 7. Saving CSV File in DataFrame

```python
df.to_csv('data.csv', index = False)
```
