```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
student = ['Ram', 'Raj', 'Rahul','Rohan', 'Rohini', 'Rajesh', 'Ashish']
marks   = [67, 87, 77, 84, 89, 88, 58]
```

### Level - 1

```python
plt.bar(student,marks)
plt.title('Student')
plt.show()
```

### Level - 2

```python
plt.bar(student,marks, width = 0.45)
plt.show()
```

### Level - 3 (colour settings)

```python
plt.bar(student,marks, color = 'g')

plt.show()
```

### Level 4 (Title Settings)

```python
plt.bar(student,marks)
plt.xlabel('Names')
plt.ylabel('Scored marks')
plt.title('comparing student marks')
plt.show()
```

### level - 6 (Horizontal Graph)

```python
student = ['Ram', 'Raj', 'Rahul','Rohan', 'Rohini', 'Rajesh', 'Ashish']
marks   = [67, 87, 77, 84, 89, 88, 58]
plt.barh(student,marks)


plt.xlabel('Marks')
plt.ylabel('Names')
plt.title('comparing marks of students')
plt.show()
```

### Level - 7 (Stacked Bar graph)

```python
class_ = ['First','Second','Third','Fourth','Fifth','Sixth','Seventh','Eighth','Ninth','Tenth']
boys = [40,32,23,43,33,44,23,44,23,12]
girls = [23,35,12,22,62,34,22,19,23,18]
```

```python
plt.bar(class_,boys,color = 'blue')
plt.bar(class_,girls,bottom = boys, color = 'orange')

plt.show()
```

### level - 8 (boys to girls ratio)

```python
f1 = []
f2 = []
for i in range(len(boys)):
    f1.append(int(girls[i]/(boys[i]+girls[i])*100))
    f2.append(int((girls[i]/boys[i])*100))
```

```python
plt.plot(class_,f1)
```

```python
plt.plot(class_,f2)
```

### Level - 9 (Stacked Bar Graph)

```python
name = ['Ashish', 'Raj', 'Rohal','Rahil']
eng = np.array([45,35,44,67])
hin = np.array([23,46,78,65])
mth = np.array([77,55,66,56])
sci = np.array([66,54,55,43])
```

```python
plt.bar(name,sci, color = 'g')
plt.bar(name,mth,bottom = sci, color ='b')
plt.bar(name,hin,bottom = sci + mth, color = 'black')
plt.bar(name,eng,bottom = sci + mth + hin, color = 'orange')

plt.title('Comparing student marks in diffrent subjects')
plt.xlabel("Names")
plt.ylabel('Marks')
plt.legend(['sci','mth','hin','eng'])
plt.show()
```
