### 4) Automate for average price

```python
for name in df['content_rating'].unique():
    lst = []
    for i in df.values:
        if(i[8] == name):
            lst.append([i[0],i[7]])
    df_gr = pd.DataFrame(lst,columns = ['name','price'])
    
    print(name,':',int(df_gr['price'].mean()))
```

### 5) Average price of application in each category

```python
for name in df['category'].unique():
    lst = []
    for i in df.values:
        if(i[1] == name):
            lst.append([i[0],i[7]])
    df_gr = pd.DataFrame(lst,columns = ['category','price'])
    
    print(name,':',int(df_gr['price'].mean()))
```

### 6) Sort the catergories based on their average price of application

```python
data = []
for name in df['category'].unique():
    lst = []
    for i in df.values:
        if(i[1] == name):
            lst.append([i[0],i[7]])
    df_gr = pd.DataFrame(lst,columns = ['category','price'])
    
    data.append([name,int(df_gr['price'].mean())])
```

```python
df_gr = pd.DataFrame(data, columns = ['name','average price'])
```

```python
df_gr.sort_values(by = 'average price', ascending = False).head(10)
```

### 7) Ratio of Free vs Paid apps in Each content rating

```python
for name in df['content_rating'].unique():
    f = 0
    p = 0
    for i in df.values:
            if(i[8] == name):
                if(i[5] == 'Free'):
                    f += 1
                else:
                    p += 1
    print(name,': ',int(f/(f+p)*100))
```

### 8) Ratio of Free vs paid app in each category

```python
lst = []
for name in df['category'].unique():
    f = 0
    p = 0
 
    for i in df.values:
            if(i[1] == name):
                if(i[5] == 'Free'):
                    f += 1
                else:
                    p += 1
    lst.append([name,int(f/(f+p)*100)])

df_pr = pd.DataFrame(lst, columns = ['name','Free%'])
```

```python
df_pr.sort_values(by = 'Free%', ascending = False).head(5)
```

### 9) Frequency of apps in each rating

```python
df['rating'].unique()
```

```python
lst = []
for rating in df['rating'].unique():
    c = 0
    for i in df.values:
        if(i[2] == rating):
            c += 1
    lst.append([rating,c])

df_pr = pd.DataFrame(lst,columns = ['rating','freq'])
```

```python
df_pr.sort_values(by = 'rating').plot(kind = 'bar')
```
