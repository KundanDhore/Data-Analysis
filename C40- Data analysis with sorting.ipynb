### Creating dataset with unique name of apps

```python
fd = pd.read_csv('googleplaystore.csv')
```

```python
fd
```

```python
fd.shape
```

```python
fd.values
```

```python
dct = {}

for i in fd.values:
  
    dct[i[0]] = {'category':i[1], 'rating' :i[2], 'review' :i[3],'size' : i[4], 'type' : i[5], 'installs' : i[6], 'price' :i[7], 'content_rating' : i[8]}
    
    
```

```python
lst = []

for name in dct.keys():
    lst.append([name, dct[name]['category'],dct[name]['rating'],dct[name]['review'],
         dct[name]['size'],dct[name]['type'],dct[name]['installs'],dct[name]['price'],dct[name]['content_rating']])
```

```python
len(lst)
```

```python
df_pr = pd.DataFrame(lst,columns = ['app','category','rating','review','size','type','installs','price','content_rating'])
```
