```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

### Level- 1 (Basic plot with small data)

```python
marks = [65,77,45,56,55,43,67,56,87,54,43,68,54,88,55,77,87,44,90,88]
```

```python
fig, ax = plt.subplots()
ax.hist(marks, bins = 5)
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Marks Distribution')
plt.show()
```

### Level - 2 (plots with large data)

```python
data = []
for i in range(10000):
    data.append(np.random.random())
fig,axs = plt.subplots(figsize = (10,5))

axs.hist(data,bins =100)
plt.show()
```

### level - 3 (normal distribution)

```python
data = np.random.randn(10000000)

fig,axs = plt.subplots(figsize = (10,5))

axs.hist(data, bins = 1000)
plt.show()
```

### level - 4 (distribution in specific range)

```python
data = np.random.randn(10000000)
data = (data + 4)*12.5

fig,axs = plt.subplots(figsize = (10,5))

axs.hist(data, bins = 1000)
plt.show()
```

### level 5 - (Normal distribution on Google Playstore dataset)

```python
df = pd.read_csv('playstore.csv')
```

```python
df.head()
```

```python
data =list(df['rating'])

fig, axs = plt.subplots()
axs.hist(data,bins = 30)
plt.show()
```

### Level - 6 (data distribution of ratings) - scratch

```python
data =list(df['rating'])
ratings = []

for rating in set(data):
    c = 0
    for i in range(len(data)):
        if(data[i] == rating):
            c+= 1
    ratings.append([rating,c])
    df_rating = pd.DataFrame(ratings, columns = ['rating','freq'])
```

```python
rating = list(df_rating.sort_values(by = 'rating')['rating'])
freq = list(df_rating.sort_values(by = 'rating')['freq'])
```

```python
plt.bar(rating,freq)
plt.show()
```

### Level 7 - Making it to normally distributed [rating>=3.5]

```python
data = []
for i in list(df['rating']):
    if(i>=3.5):
        data.append(i)
        
fig, axs = plt.subplots()
axs.hist(data,bins = 15)
plt.show()
```

### Level 8 - Making it to normally distributed [rating>=3.5 and rating < 5]

```python
data = []
for i in list(df['rating']):
    if(i>=3.5 and i <5.0):
        data.append(i)
        
fig, axs = plt.subplots()
axs.hist(data,bins = 15)
plt.show()
```
